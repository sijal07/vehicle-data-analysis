```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("electric_vehicles_spec_2025.csv")
```

```python
df.head()
```

```python
df.drop(['source_url','model', 'brand','segment','cargo_volume_l'], axis=1, inplace=True)

```

```python
cat= ['battery_type','drivetrain', 'car_body_type', 'fast_charge_port']
for i in cat:
    df[i] = LabelEncoder().fit_transform(df[i])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['number_of_cells'].fillna(df['number_of_cells'].mean(), inplace=True)
df['torque_nm'].fillna(df['torque_nm'].mean(), inplace=True)
df['fast_charging_power_kw_dc'].fillna(df['fast_charging_power_kw_dc'].mean(), inplace=True)
df['towing_capacity_kg'].fillna(df['towing_capacity_kg'].mean(), inplace=True)


```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.drop('efficiency_wh_per_km', axis=1)
y = df['efficiency_wh_per_km']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
# Decision Tree
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_train, y_train)
# Random Forest
forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(X_train, y_train)
# Predict using training data
lin_preds = lin_reg.predict(X_train)
tree_preds = tree_reg.predict(X_train)
forest_preds = forest_reg.predict(X_train)
```

```python
from sklearn.metrics import r2_score
lin=r2_score(y_train, lin_preds)
```

```python
from sklearn.metrics import r2_score
tree= r2_score(y_train, tree_preds)
```

```python
from sklearn.metrics import r2_score
forest= r2_score(y_train, forest_preds)
```

```python
print("Linear Regression :", lin)
print("Decision Tree :", tree)
print("Random Forest :", forest)
```

```python
dataset = pd.read_csv('new_data.csv')
a = dataset.drop('drivetrain', axis=1)
b = dataset['drivetrain']
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
a_train= sc.fit_transform(a_train)
a_test= sc.fit_transform(a_test)


```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
a_train = pca.fit_transform(a_train)
a_test= pca.fit_transform(a_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(a_train,b_train)
```

```python
b_pred = classifier.predict(a_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(b_test, b_pred)
print(cm)
accuracy_score(b_test, b_pred)
```

```python
import seaborn as sns

accuracy = accuracy_score(b_test, b_pred)
print(f"Accuracy: {accuracy:.2f}")

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=["zero", "one", "two"],
            yticklabels=["zero", "one", "two"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Logistic Regression Confusion Matrix\nAccuracy: {accuracy:.2f}')
plt.tight_layout()
plt.show()
```

```python
from matplotlib.colors import ListedColormap


X_set, y_set = a_train, b_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red', 'green', 'blue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('red', 'green', 'blue'))(i), label=j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('PCI')
plt.ylabel('PCZ')
plt.legend()
plt.show()
```
